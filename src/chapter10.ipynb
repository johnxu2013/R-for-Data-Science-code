```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
mpg = pd.read_csv("data/mpg.csv")

# 1. Scatter plot with color
sns.scatterplot(data=mpg, x='displ', y='hwy', hue='class')
plt.show()

# 2. Scatter plot with shape
sns.scatterplot(data=mpg, x='displ', y='hwy', style='class')
plt.show()

# 3. Scatter plot with size - Note: using size for a discrete variable is not advised
sns.scatterplot(data=mpg, x='displ', y='hwy', size='class')
plt.show()

# 4. Scatter plot with alpha - Note: seaborn does not directly support alpha by categories
# ... This might require a manual loop or using another library such as matplotlib

# 5. Scatter plot with fixed color
sns.scatterplot(data=mpg, x='displ', y='hwy', color='blue')
plt.show()

# 6. Basic scatter plot
sns.scatterplot(data=mpg, x='displ', y='hwy')
plt.show()

# 7. Scatter with regression line
sns.regplot(data=mpg, x='displ', y='hwy')
plt.show()
```

```python
# Bar plot of diamonds data by clarity
diamonds['clarity'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Clarity')
plt.ylabel('Count')
plt.title('Bar plot of diamonds by clarity')
plt.show()
```

```python
# Bar plot of diamonds data by clarity, flipped
diamonds['clarity'].value_counts().sort_index().plot(kind='barh', color='skyblue', edgecolor='black')
plt.ylabel('Clarity')
plt.xlabel('Count')
plt.title('Horizontal bar plot of diamonds by clarity')
plt.show()
```

```python
# Calculate the counts of each clarity value
clarity_counts = diamonds['clarity'].value_counts().sort_index()

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# Create a polar bar plot
angles = [i / float(clarity_counts.size) * 2 * 3.141592653589793 for i in range(clarity_counts.size)]
ax.bar(angles, clarity_counts, width=0.4, color='skyblue', edgecolor='black', alpha=0.7)

# Set the theta tick labels
ax.set_xticks(angles)
ax.set_xticklabels(clarity_counts.index)

plt.title('Polar bar plot of diamonds by clarity')
plt.show()
```
