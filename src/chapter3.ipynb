```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Arithmetic operations
print(1 / 200 * 30)
print((59 + 73 + 2) / 3)
print(math.sin(math.pi / 2))

# Variable assignment
x = 3 * 4
print(x)

primes = np.array([2, 3, 5, 7, 11, 13])
print(primes * 2)
print(primes - 1)

this_is_a_really_long_name = 2.5

r_rocks = 2 ** 3
print(r_rocks)

# Sequences
print(np.arange(1, 11))

x = "hello world"
print(x)

my_variable = 10
print(my_variable)

# Loading and plotting mpg dataset (using seaborn's load_dataset for demonstration purposes)
mpg = sns.load_dataset('mpg')



# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=mpg, x='displacement', y='mpg')
sns.regplot(data=mpg, x='displacement', y='mpg', scatter=False, ci=None)
plt.show()

# For information on the mpg dataset
print(mpg.info())

# List all datasets available in seaborn (as a substitute for ls("package:tidyverse"))
print(sns.get_dataset_names())

# Bar plot and scatter plot
# plt.figure(figsize=(10, 6))
# sns.countplot(data=mpg, x='class')
# plt.show()

# plt.figure(figsize=(10, 6))
# sns.scatterplot(data=mpg, x='city_mpg', y='highway_mpg')
# plt.show()

# Save the bar plot
# plt.figure(figsize=(10, 6))
# sns.countplot(data=mpg, x='class')
# plt.savefig("mpg-plot.png")
```
