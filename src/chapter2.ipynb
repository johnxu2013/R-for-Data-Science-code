```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading data
penguins = sns.load_dataset('penguins')

# Displaying the first few rows of the dataframe
print(penguins.head())

# Basic scatter plot
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g')
plt.show()

# Scatter plot with species color
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species')
plt.show()

# Scatter plot with regression line
sns.lmplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species')
plt.show()

# Scatter plot with custom labels and colorblind palette
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species', style='species')
plt.title('Body mass and flipper length')
plt.xlabel('Flipper length (mm)')
plt.ylabel('Body mass (g)')
plt.legend(title='Species')
plt.show()

# Get dimensions of the dataframe
print(f"Number of rows: {len(penguins)}")
print(f"Number of columns: {len(penguins.columns)}")

# Scatter plots for bill depth and length
sns.scatterplot(data=penguins, x='bill_depth_mm', y='bill_length_mm', hue='species')
plt.show()

# Violin plot for bill depth by species
sns.violinplot(data=penguins, x='species', y='bill_depth_mm')
plt.show()

# Scatter plot with island color
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='island')
plt.show()

# Bar plot for species
sns.countplot(data=penguins, x='species')
plt.show()

# Histogram for body mass with different bin sizes
sns.histplot(data=penguins, x='body_mass_g', bins=20)
plt.show()

sns.histplot(data=penguins, x='body_mass_g', bins=2000)
plt.show()

# Density plot for body mass
sns.kdeplot(data=penguins, x='body_mass_g')
plt.show()

# Bar plot with color
sns.countplot(data=penguins, x='species', color='red')
plt.show()

# Boxplot of body mass by species
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.show()

# Density plot with species color and fill
sns.kdeplot(data=penguins, x='body_mass_g', hue='species', fill=True, alpha=0.5)
plt.show()

# Bar plots with fill by species or island
sns.countplot(data=penguins, x='island', hue='species')
plt.show()

sns.countplot(data=penguins, x='species', hue='island')
plt.show()

# Save a plot to file
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species', style='island')
plt.savefig("penguin-plot.png")

# You can use similar methods to plot and save figures for the `mpg` dataset, which is also available in seaborn
```
