```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you've loaded the nycflights13 dataset into a DataFrame called flights
flights = pd.read_csv('flights.csv')

# Filter rows where dep_delay and arr_delay are not NaN
not_cancelled = flights.dropna(subset=['dep_delay', 'arr_delay'])

# Group by year, month, day and calculate the mean of dep_delay
grouped = not_cancelled.groupby(['year', 'month', 'day']).agg(mean_dep_delay=('dep_delay', 'mean')).reset_index()

# Check if a value is NaN
is_na = pd.isna(3)  # This will return False

# Get current working directory
import os
cwd = os.getcwd()
print(cwd)

# Plotting using seaborn and saving the plot
diamonds = pd.read_csv('data/diamonds.csv')
# Create a hexbin plot
plt.figure(figsize=(10, 6))
plt.hexbin(diamonds['carat'], diamonds['price'], gridsize=50, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Hexbin Plot of Carat vs. Price')
plt.show()

plt.savefig("data/diamonds.png")

```
